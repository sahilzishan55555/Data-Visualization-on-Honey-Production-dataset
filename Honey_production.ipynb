# 3. Display the percentage distribution of the data in each year using the pie chart.

```python
total_prodvalue_by_year = dataframe.groupby('year')['prodvalue'].sum()
year= dataframe["year"].unique()
x= total_prodvalue_by_year
y= year
plt.pie(x,labels=y,autopct='%1.1f%%')
plt.show()
```

# 4. Plot and Understand the distribution of the variable "price per lb" using displot, and write your findings.

```python
priceperLb = dataframe["priceperlb"]     #price per lb around 1.5 occures maximum times and about 100 times
sns.displot(priceperLb)                  #The most common price is 1.5
plt.title("Distribuation of price per LB")#price per Lb around 4.0 occoures minimum times
plt.xlabel("Price per Lb")                #The least commom price is 4.0
plt.ylabel("Count")                       #The average price would be around between 2 and 2.25
plt.show()
```

# 5. Plot and understand the relationship between the variables 'numcol' and 'prodval' through scatterplot, and write your findings.

```python
a = dataframe["numcol"]
b = dataframe["prodvalue"]
plt.scatter(a,b)
plt.title("Scatter plot of numcol and prodval")
plt.xlabel("numcol")
plt.ylabel("prodvalue")
plt.show()
```

_ We can simply say that there is correlation between both the variables
_ There is positive correlation between both the variables
_ That simply means that if one variable is increasing then other will also increase
_ And if one variable is decreasing then other will also decrease
_ Any impact on any one variable will definitely effect the other variable 
_They both are positively correlated 

# 6. Plot and understand the relationship between categorical variable 'year' and a numerical variable 'prodvalue' through boxplot, and write your findings.

```python
dataframe.dtypes
```

```python
plt.figure(figsize=(10, 6))
plt.boxplot(dataframe['year'],patch_artist=True,showmeans=True)
plt.xlabel('Year')
plt.ylabel('Count')
plt.title('Boxplot of Year')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.boxplot(dataframe['prodvalue'],patch_artist=True,showmeans=True)
plt.title('Boxplot of prodvalue')
plt.ylabel('prodvalue')
plt.show()
```

# 7. Visualize and understand the relationship between the multiple pairs of variables throughout different years using pairplot and add your inferences. (use columns 'numcol', 'yield percol', 'total prod', 'prodvalue','year')

```python


# Selecting the columns for the pair plot
columns = ['numcol', 'yieldpercol', 'totalprod',  'prodvalue', 'year']

# Creating a new DataFrame with the selected columns
data_for_pairplot = dataframe[columns]

# Plotting the pair plot
sns.pairplot(data_for_pairplot)

# Display the plot
plt.show()
```

# 8. Display the correlation values using a plot and add your inferences. (use columns 'numcol', 'yield percol', 'total prod', 'stocks', 'price per lb', 'prodvalue')

```python

# Selecting the columns for correlation analysis
columns = ['numcol', 'yieldpercol', 'totalprod', 'stocks', 'priceperlb', 'prodvalue']

# Creating a new DataFrame with the selected columns
data_for_correlation = dataframe[columns]

# Calculating the correlation matrix
correlation_matrix = data_for_correlation.corr()

# Plotting the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')

# Add title and display the plot
plt.title('Correlation Matrix')
plt.show()
```
